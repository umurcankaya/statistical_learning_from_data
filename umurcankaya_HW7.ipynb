```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
import scipy.stats as st
from matplotlib import cm
```

```python
def chi2dist(chi2,n):
    return 1/(2**(n/2)*gamma(n/2)) * chi2**(n/2-1) * np.exp(-chi2/2)
```

```python
f = plt.figure(figsize=(10,5))
chi2 = np.linspace(0,4,100)
for n in range(1,6):
    plt.plot(chi2, chi2dist(chi2,n), label='dof=%d'%n, lw=2)
plt.legend(fontsize=15)
plt.xlabel('$\chi^2$',fontsize=15)
plt.ylabel('$P(\chi^2,n)$',fontsize=15)
```

```python
chi2 = np.linspace(1,10,100)
n = np.linspace(1,10,100)

Chi2, N = np.meshgrid(chi2, n, indexing='xy')
c = plt.contourf(n,chi2,chi2dist(Chi2,N),origin='lower', cmap=cm.rainbow)
plt.axis('scaled')
plt.contour(c,colors='k',lw=2)
plt.colorbar(c,label='$P(\chi^2,n)$')
plt.xlabel('n')
plt.ylabel('$\chi^2$')
```

```python
np.random.seed(1000)
x = np.random.poisson(30,1000)
var = np.unique(x)
count = np.array([len(x[x==element]) for element in var])
alpha = np.sqrt(count)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
fs=15
ax.bar(var,count,yerr=alpha,color='C1',edgecolor='k',alpha=0.5)
ax.set_ylabel('count', fontsize=fs)
ax.set_xlabel('variable', fontsize=fs)
#ax.errorbar(indep, dep, yerr=alpha, fmt='o')
```
